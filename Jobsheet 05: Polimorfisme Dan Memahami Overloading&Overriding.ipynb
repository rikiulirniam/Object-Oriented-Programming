### Praktikum 01: Polimorfisme dengan Inheritance dan Overriding

```python
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print(f"{self.nama} terbang dengan cara umum.")

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara burung.")


class Elang(Burung):
    def __init__(self, nama, rentang_sayap):
        super().__init__(nama)
        self.rentang_sayap = rentang_sayap

    def terbang(self):
        print(f"{self.nama} sedang terbang tinggi melayang di angkasa")

    def bersuara(self):
        print(f"{self.nama} berteriak nyaring.")

class Pipit(Burung):
    def __init__(self, nama, warna_bulu):
        super().__init__(nama)
        self.warna_bulu = warna_bulu

    def terbang(self):
        print(f"{self.nama} sedang terbang cepat di antara pepohonan")

    def bersuara(self):
        print(f"{self.nama} berkicau merdu: Cit cit!")


def demonstrasi_aksi_burung(list_burung: list):
    print("\n Aksi burung:")
    for burung in list_burung:
        burung.terbang()
        burung.bersuara()
        print("-" * 15)

if __name__ == "__main__":
    elang_jawa = Elang("Elang Jawa", 1.5)
    pipit_gereja = Pipit("Pipit Gereja", "Coklat")
    burung_aneh = Burung("Burung Misterius")
    koleksi_burung = [elang_jawa, pipit_gereja, burung_aneh]
    demonstrasi_aksi_burung(koleksi_burung)

```

### Praktikum 02: Polimorfisme dengan Duck Typing

```python
class Kucing:
    def bersuara(self):
        print("Kucing: Meow!")

class Anjing:
    def bersuara(self):
        print("Anjing: Guk guk!")

class Bebek:
    def bersuara(self):
        print("Bebek: Kwek kwek!")

class Mobil:
    def jalankan(self):
        print("Mobil: Brummm!")

def buat_suara(list_object: list):
    try:
        list_object.bersuara()
    except AttributeError:
        print(f"{list_object.__class__.__name__} tidak bisa bersuara.")

if __name__ == "__main__":
    kucing = Kucing()
    anjing = Anjing()
    bebek = Bebek()
    mobil = Mobil()

    daftar_objek = [kucing, anjing, bebek, mobil]

    print("Demonstrasi Duck Typing:")
    for item in daftar_objek:
        buat_suara(item)
```

### Praktikum 03: Simulasi Overloading dalam Metode Kelas dengan *args

```python
class KalkulatorSederhana:
    def __init__(self, nama):
        self.nama = nama

    def tambah(self, *args: int) -> float:
        print(f"\nMemanggil metode tambah oleh {self.nama}")
        if not args:
            print("Tidak ada angka yang diberikan untuk penjumlahan.")
            return 0

        total = 0
        valid_input = True
        for angka in args:
            if not isinstance(angka, (int, float)):
                print(f"Input tidak valid: {angka} bukan angka. Semua input harus berupa angka.")
                valid_input = False
                break
            total += angka
        if not valid_input:
            print(f"Hasil penjumlahan (dengan beberapa input diabaikan): {total}")
        else :
            print(f"Hasil penjumlahan: {total}")
        return total

if   __name__ == "__main__":
    calc = KalkulatorSederhana("Calc-01")
    print("\n--- Percobaan Penjumlahan ---")
    calc.tambah(5, 10)
    calc.tambah(2, 3, 5, 10)
    calc.tambah(100)
    calc.tambah() # Tanpa argumen
    calc.tambah(1, 2, "tiga", 4, 5.5) # Dengan input non-angka
```

### Praktikum 04: Simulasi Overloading dalam Metode Kelas dengan *args dan **kwargs

```python
class DataProcessor:
    def __init__(self, nama):
        self.nama = nama

    def process(self, *args, **kwargs):
        if args:
            print("Argumen Posisi Diterima:")
            for i, arg in enumerate(args):
                print(f" args[{i}]: {arg} (tipe: {type(arg).__name__ })")
        else:
                print("Tidak ada argumen posisi.")

        if kwargs:
            print("\nArgumen Kata Kunci Diterima:")
            for key, value in kwargs.items():
                print(f" {key}: {value} (tipe: {type(value).__name__})")
        else:
            print("Tidak ada argumen kata kunci.")
            print("--"*15)

if __name__ == "__main__":
    processor1 = DataProcessor("DP-001")

    print("\nPanggilan 1: Tanpa argumen tambahan")
    processor1.process()

    print("\nPanggilan 2: Hanya argumen posisi")
    processor1.process(100, "Status OK", 99.9, False)

    print("\nPanggilan 3: Hanya argumen kata kunci")
    processor1.process(nama="John", usia=30, kota="New York")

    print("\nPanggilan 4: Kombinasi argumen posisi dan kata kunci")
    processor1.process("Task-A", "Task-B", status="Running", priority="High", thread_id=54321)

    print("\nPanggilan 5: Argumen posisi kompleks")
    list_ids = ["A1", "B2", "C3"]
    dict_params = {"timeout": 30, "retry": 3}
    processor1.process(list_ids, dict_params, owner="system", enabled=True)
```

### Praktikum 05: Kombinasi Polimorfisme (Inheritance & Duck Typing)

```python
import math

class Bentuk:
    def nama_bentuk(self)->str:
        return "Bentuk Generik"
    def hitung_luas(self):
        raise NotImplementedError("Metode hitung_luas() harus diimplementasikan oleh subclass.")

class Lingkaran(Bentuk):
    def __init__(self, radius):
        self.radius = radius

    def nama_bentuk(self) -> str:
        return "Lingkaran"

    def hitung_luas(self):
        return math.pi * (self.radius ** 2)

class Persegi(Bentuk):
    def __init__(self, sisi):
        self.sisi = sisi

    def nama_bentuk(self) -> str:
        return "Persegi"

    def hitung_luas(self):
        return self.sisi * self.sisi

class TaplakMeja:
    def __init__(self, panjang, lebar):
        self.panjang = panjang
        self.lebar = lebar
    def hitung_luas(self):
        return self.panjang * self.lebar
    def deskripsi(self):
        return f"Taplak Meja  {self.panjang} x {self.lebar}"

def tampilkan_info_luas(object_dgn_luas):
    try:
        luas = object_dgn_luas.hitung_luas()
        try:
            nama = object_dgn_luas.nama_bentuk()
            print(f"{nama} memiliki luas: {luas:.2f}")
        except AttributeError:
            nama = type(object_dgn_luas).__name__
        print(f"{nama} memiliki luas: {luas:.2f}")
    except AttributeError:
        print(f"Objek {type(object_dgn_luas).__name__} tidak memiliki metode hitung_luas().")
    except NotImplementedError:
        print(f"Objek {type(object_dgn_luas).__name__} belum mengimplementasikan metode hitung_luas().")
if __name__ == "__main__":
    lingkaran1 = Lingkaran(7)
    persegi1 = Persegi(5)
    taplak1 = TaplakMeja(1.5, 0.8)
    bentuk_dasar = Bentuk()

    string_biasa = "ini string"

    daftar_item = [lingkaran1, persegi1, taplak1, bentuk_dasar, string_biasa]

    print("Menampilkan info luas (Polymorphism campuran)")
    for item in daftar_item:
        tampilkan_info_luas(item)
```

### Praktikum 06: Kontrol Perilaku Polimorfik dengan isinstance()

```python
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print(f"{self.nama} terbang dengan cara umum.")

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara burung.")

class Elang(Burung):
    def __init__(self, nama, rentang_sayap):
        super().__init__(nama)
        self.rentang_sayap = rentang_sayap

    def terbang(self):
        print(f"{self.nama} sedang terbang tinggi melayang di angkasa")

    def bersuara(self):
        print(f"{self.nama} berteriak nyaring.")

    def berburu(self):
        print(f"{self.nama} sedang berburu mangsa di hutan.")

class Pipit(Burung):
    def __init__(self, nama, warna_bulu):
        super().__init__(nama)
        self.warna_bulu = warna_bulu

    def terbang(self):
        print(f"{self.nama} sedang terbang cepat di antara pepohonan")

    def bersuara(self):
        print(f"{self.nama} berkicau merdu: Cit cit!")

    def membangun_sarang(self):
        print(f"{self.nama} sedang membangun sarang di dahan pohon.")

def interaksi_burung(burung):
    print(f"\n--- Interaksi dengan {type(burung).__name__ }: {getattr(burung, 'nama', 'Objek tidak dikenal')} ---")

    if isinstance(burung, Burung):
        burung.terbang()
        burung.bersuara()
        if isinstance(burung,Elang):
            burung.berburu()
        elif isinstance(burung, Pipit):
            burung.membangun_sarang()
        else:
            print(f"{burung.nama} adalah burung umum, selain Elang dan Pipit.")
    else:
        print(f"Objek {type(burung).__name__} bukanlah burung.")

if __name__ == "__main__":
    elang_sumatra = Elang("Elang Sumatra", 1.8)
    pipit_rumah = Pipit("Pipit Rumah", "Abu-abu")
    merak = Burung("Merak")
    kucing_tetangga = "Meng"

    koleksi_makhluk	=	[elang_sumatra,	pipit_rumah,	merak, kucing_tetangga]

    for makhluk in koleksi_makhluk:
        interaksi_burung(makhluk)

```

## Penugasan

```python
class Komputer:
    def __init__(self, merk: str, processor: str, ram_gb: int):
        self.merk = merk
        self.processor = processor
        self.ram_gb = ram_gb

    def info_spesifikasi(self):
        print(f" Merk {self.merk}:")
        print(f" Processor: {self.processor}")
        print(f" RAM: {self.ram_gb} GB")

    def jalankan_aplikasi(self, nama_aplikasi: str):
        print(f"{self.merk} menjalankan aplikasi {nama_aplikasi}...")

    def upgrade_ram(self, tambahan_gb: int, tipe_ram: str = "DDR4"):
        if tambahan_gb > 0:
            self.ram_gb += tambahan_gb
            print(f"RAM {self.merk} diupgrade sebesar {tambahan_gb} GB (Tipe: {tipe_ram})]). Total RAM sekarang: {self.ram_gb} GB. ")
        else:
            print("Tambahan RAM harus lebih besar dari 0.")

class Laptop(Komputer):
    def __init__(self, merk: str, processor: str, ram_gb: int, ukuran_layar_inch: float, berat_kg: float):
        super().__init__(merk, processor, ram_gb)
        self.ukuran_layar_inch = ukuran_layar_inch
        self.berat_kg = berat_kg

    def info_spesifikasi(self):
        super().info_spesifikasi()
        print(f" Ukuran Layar: {self.ukuran_layar_inch} inch")
        print(f" Berat: {self.berat_kg} kg")

class Desktop(Komputer):
    def __init__(self, merk: str, processor: str, ram_gb: int, jenis_casing: str, monitor_external: bool):
        super().__init__(merk, processor, ram_gb)
        self.jenis_casing = jenis_casing
        self.monitor_external = monitor_external

    def info_spesifikasi(self):
        super().info_spesifikasi()
        print(f" Jenis Casing: {self.jenis_casing}")
        print(f" Monitor External: { 'Ya' if self.monitor_external else 'Tidak'}")

def cetak_semua_spesifikasi(daftar_komputer: list):
    print("\n--- Informasi Spesifikasi Komputer ---")
    for komputer in daftar_komputer:
        if isinstance(komputer, Komputer):
            komputer.info_spesifikasi()

            print("-" * 30)
        else:
            print(f"Objek {type(komputer).__name__} bukanlah komputer.")

if __name__ == "__main__":
    laptop_object = Laptop("Dell XPS 13", "Intel Core i7", 16, 13.3, 1.2)
    desktop_object = Desktop("HP Pavilion", "AMD Ryzen 5", 32, "Mid Tower", True)
    benda_aneh = "Bukan Komputer"

    list_computer = [laptop_object, desktop_object, benda_aneh]

    cetak_semua_spesifikasi(list_computer)

    laptop_object.upgrade_ram(8, "DDR5")
```
