### Praktikum 01: Inheritance Dasar

```python
class Kendaraan:
    def __init__(self, merk:str):
        self.merk = merk

    def mulai_mesin(self):
        print(f"{self.merk} mesin dimulai.")


class Mobil(Kendaraan):
    def __init__(self, merk:str, warna:str):
        self.merk = merk
        self.warna = warna

    def info_mobil(self):
        print(f"Mobil {self.merk} berwarna {self.warna}.")

if __name__ == "__main__":
    mobil_tesla = Mobil("Tesla", "Merah")
    mobil_tesla.info_mobil()
    mobil_tesla.mulai_mesin()

    print(f"Merk mobil: {mobil_tesla.merk}")

```

## Praktikum 02: Menggunakan super() dalam konstruktor

```python
class Person:
    def __init__(self, nama:str, usia:int):
        self.nama = nama
        self.usia = usia

    def perkenalkan_diri(self):
        print(f"Halo, nama saya {self.nama}, usia saya {self.usia} tahun.")

class Student(Person):
    def __init__(self, nama:str, usia:int, student_id:int, jurusan:str):
        super().__init__(nama, usia)
        self.student_id = student_id
        self.jurusan = jurusan
        print(f"Inisialisasi attribut Student selesai untuk {nama}")

    def info_akademik(self):
        print(f"ID Mahasiswa: {self.student_id}")
        print(f"Jurusan: {self.jurusan}")

    def perkenalkan_diri(self):
        super().perkenalkan_diri()
        print(f"Saya adalah mahasiswa dengan ID {self.student_id}, jurusan {self.jurusan}.")

if __name__ == "__main__":
    dosen = Person("Pak Anton", 45)
    print("-" * 20)
    dosen.perkenalkan_diri()
    print("\n" + "=" * 30 + "\n")
    mahasiswa = Student("Dewi", 20, "MHS001", "Teknik Komputer")
    print("-" * 20)
    mahasiswa.perkenalkan_diri()
    print("-" * 20)
    print(f"\nUsia mahasiswa {mahasiswa.nama}: {mahasiswa.usia}")
```

## Praktikum 3: Method Overriding dan super()

```python
class Hewan:
    def __init__(self, nama: str):
        self.nama = nama

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara generik.")
    def tampilkan_info(self):
        print(f"Hewan ini bernama {self.nama}.")

class Kucing(Hewan):
    def __init__(self, nama: str, ras:str):
        super().__init__(nama)
        self.ras = ras

    def bersuara(self):
        print(f"{self.nama} (Kucing) mengeong: Meow!")
    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Ras kucing: {self.ras}")

if __name__ == "__main__":
    hewan_umum = Hewan("Makhluk")
    kucing_persia = Kucing("Puspus", "Persia")

    print("Info Hewan Umum:")
    hewan_umum.tampilkan_info()
    hewan_umum.bersuara()
    print("-" * 20)

    print("Info Kucing Persia:")
    kucing_persia.tampilkan_info() # Akan memanggil versi Kucing
    kucing_persia.bersuara() # Akan memanggil versi Kucing
```

## Praktikum 4: Multi-level Inheritance (Pewarisan Bertingkat)

```python
class Organisme:
    def __init__(self, nama:str):
        self.nama = nama
        print(f"Organisme '{self.nama}' diciptakan.")

    def bernapas(self):
        print(f"{self.nama} sedang bernafas.")

class Hewan(Organisme):
    def __init__(self, nama:str, jenis_makanan:str):
        super().__init__(nama)
        self.jenis_makanan = jenis_makanan

    def bergerak(self):
        print(f"{self.nama} sedang bergerak.")

class Mamalia(Hewan):
    def __init__(self, nama:str, jenis_makanan:str, jumlah_kaki:int):
        super().__init__(nama, jenis_makanan)
        self.jumlah_kaki = jumlah_kaki
        print(f"Mamalia '{self.nama}' memiliki {self.jumlah_kaki} kaki.")

    def menyusui(self):
        print(f"{self.nama} sedang menyusui.")

if __name__ == "__main__":
    kucing = Mamalia("Kucing Persia", "Karnivora", 4)
    print("-" * 20)

    # Memanggil metode dari kelas Mamalia
    kucing.menyusui()

    # Memanggil metode yang diwarisi dari kelas Hewan
    kucing.bergerak()
    print(f"Jenis makanan: {kucing.jenis_makanan}")

    # Memanggil metode yang diwarisi dari kelas Organisme
    kucing.bernapas()
    print(f"Nama organisme: {kucing.nama}")
```

## Praktikum 5: Hierarchical Inheritance (Pewarisan Hierarkis)

```python
import locale

try:
    locale.setlocale(locale.LC_ALL, 'id_ID.UTF-8')
except locale.Error:
    print("Locale id_ID.UTF-8 tidak tersedia, menggunakan locale default.")

def format_rupiah(angka):
    return locale.currency(angka, grouping=True, symbol='Rp ')

class Pegawai:
    def __init__ (self, nama:str, id_pegawai:str, gaji_pokok:float):
        self.nama = nama
        self.id_pegawai = id_pegawai
        self.gaji_pokok = gaji_pokok

    def hitung_gaji(self) -> float:
        return self.gaji_pokok

    def tampilkan_info(self):
        print(f"ID: {self.id_pegawai}, Nama: {self.nama}")
        print(f"Gaji Pokok: {format_rupiah(self.gaji_pokok)}")

class Manager(Pegawai):
    def __init__(self, nama:str, id_pegawai:str, gaji_pokok:float, tunjangan_jabatan:float):
        super().__init__(nama, id_pegawai, gaji_pokok)
        self.tunjangan_jabatan = tunjangan_jabatan

    def hitung_gaji(self) -> float:
        return super().hitung_gaji() + self.tunjangan_jabatan

    def tampilkan_info(self):
        print("--- Info Manager ---")
        super().tampilkan_info()
        print(f"Tunjangan: {format_rupiah(self.tunjangan_jabatan)}")
        print(f"Total Gaji: {format_rupiah(self.hitung_gaji())}")

class StafTeknis(Pegawai):
    def __init__(self, nama:str, id_pegawai:str, gaji_pokok:float, keahlian: str, bonus_keahlian: float):
        super().__init__(nama, id_pegawai, gaji_pokok)
        self.keahlian = keahlian
        self.bonus_keahlian = bonus_keahlian

    def hitung_gaji(self)->float:
        return super().hitung_gaji() + self.bonus_keahlian

    def tampilkan_info(self):
        print("--- Info Staf Teknis ---")
        super().tampilkan_info()
        print(f"Keahlian: {self.keahlian}")
        print(f"Bonus Keahlian: {format_rupiah(self.bonus_keahlian)}")
        print(f"Total Gaji: {format_rupiah(self.hitung_gaji())}")

if __name__ == "__main__":
    manager1 = Manager("Budi Santoso", "M001", 10000000, 5000000)
    staf1 = StafTeknis("Citra Lestari", "S001", 7000000, "Python Programming", 1500000)
    pegawai_baru = Pegawai("Rian", "P005", 5000000) # Pegawai biasa

    print("Menampilkan Informasi Pegawai:")
    manager1.tampilkan_info()
    print("-" * 30)
    staf1.tampilkan_info()
    print("-" * 30)

    print("--- Info Pegawai Baru ---")
    pegawai_baru.tampilkan_info()
    print(f" Total Gaji: {format_rupiah(pegawai_baru.hitung_gaji())}") # Memanggil versi Pegawai
    print("-" * 30)
```

## Praktikum 6: Memeriksa Tipe Instance dengan isinstance()

```python
class BentukGeometri:
    def __init__(self, nama:str):
        self.nama = nama

    def tampilkan_nama(self):
        print(f"Nama Bentuk: {self.nama}")

    def hitung_luas(self):
        print(f"Luas {self.nama} tidak dapat dihitung (generik).")
class PersegiPanjang(BentukGeometri):
    def __init__(self, nama:str, panjang:float, lebar:float):
        super().__init__(nama)
        self.panjang = panjang
        self.lebar = lebar

    def hitung_luas(self) -> float:
        return self.panjang * self.lebar

class Segitiga(BentukGeometri):
    def __init__(self, nama:str, alas:float, tinggi:float):
        super().__init__(nama)
        self.alas = alas
        self.tinggi = tinggi

    def hitung_luas(self) -> float:
        return 0.5 * self.alas * self.tinggi

def proses_bentuk(bentuk):
    print(f"\nMemproses objek: {type(bentuk)}")

    if isinstance(bentuk, BentukGeometri):
        bentuk.tampilkan_nama()
        print("Objek ini adalah instance dari BentukGeometri (atau turunannya).")

    if isinstance(bentuk, PersegiPanjang):
        print("-> Objek ini juga instance dari PersegiPanjang.")
    elif isinstance(bentuk, Segitiga):
        print("-> Objek ini juga instance dari Segitiga.")
    else:
        print("Objek ini BUKAN instance dari BentukGeometri.")

    if isinstance(bentuk, str):
        print("Objek ini adalah string.")


if __name__ == "__main__":
    pp = PersegiPanjang("Persegi Panjang", 4, 3)
    seg = Segitiga("Segitiga", 5, 2)
    bg = BentukGeometri("Lingkaran")
    teks = "Ini String"

    daftar_objek = [pp, seg, bg, teks]

    for obj in daftar_objek:
        proses_bentuk(obj)
```

## Penugasan

```python
# Kelas Induk
class Mahasiswa:
    def __init__(self, nama, nim, ipk):
        self.nama = nama
        self.nim = nim
        self.ipk = ipk

    def tampilkan_info(self):
        print(f"Nama: {self.nama}")
        print(f"NIM: {self.nim}")
        print(f"IPK: {self.ipk}")

    def hitung_predikat(self):
        if self.ipk >= 3.5:
            print("Dengan Pujian")
        elif self.ipk >= 3.0:
            print("Sangat Memuaskan")
        elif self.ipk >= 2.5:
            print("Memuaskan")
        else:
            print("Cukup")

# Kelas Anak 1 - Mahasiswa Sarjana
class MahasiswaSarjana(Mahasiswa):
    def __init__(self, nama, nim, ipk, semester, sks_lulus):
        super().__init__(nama, nim, ipk)
        self.semester = semester
        self.sks_lulus = sks_lulus

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Semester: {self.semester}")
        print(f"SKS Lulus: {self.sks_lulus}")

# Kelas Anak 2 - Mahasiswa Magister
class MahasiswaMagister(Mahasiswa):
    def __init__(self, nama, nim, ipk, judul_tesis, nama_pembimbing):
        super().__init__(nama, nim, ipk)
        self.judul_tesis = judul_tesis
        self.nama_pembimbing = nama_pembimbing

    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Judul Tesis: {self.judul_tesis}")
        print(f"Pembimbing: {self.nama_pembimbing}")

    # Override predikat
    def hitung_predikat(self):
        if self.ipk >= 3.7:
            print("Lulus dengan Pujian Magister")
        elif self.ipk >= 3.3:
            print("Sangat Memuaskan")
        else:
            print("Memuaskan")

if __name__=="__main__":
    print("=== Mahasiswa Baru ===")

mhs_baru = Mahasiswa("Jouhan Johari", "202531423", 3.83)
mhs_baru.tampilkan_info()
mhs_baru.hitung_predikat()

print("\n=== Mahasiswa Sarjana ===")
mhs_s1 = MahasiswaSarjana("Muhammad Ramadhan Alfian Nurunnazla", "2024160634", 3.4, semester=2, sks_lulus=23)
mhs_s1.tampilkan_info()
mhs_s1.hitung_predikat()

print("\n=== Mahasiswa Sarjana 2 ===")
mhs_s1b = MahasiswaSarjana("Risky Salisa Rahma", "20231456732", 2.3, semester=4, sks_lulus=54)
mhs_s1b.tampilkan_info()
mhs_s1b.hitung_predikat()

print("\n=== Mahasiswa Magister ===")
mhs_s2 = MahasiswaMagister("Intan Permata Putri", "2021465124", 3.8, "Studi Keamanan Sistem IOT Berbasis MQTT", "Dr. Yustinus Eko")
mhs_s2.tampilkan_info()
mhs_s2.hitung_predikat()

print("\n=== Mahasiswa Magister 2 ===")
mhs_s2b = MahasiswaMagister("Muhammad Ali Irfan", "201927512", 3.2, "Optimasi Penggunaan Energi Pada Smart Home", "Prof. Bens")
mhs_s2b.tampilkan_info()
mhs_s2b.hitung_predikat()
```
