### Praktikum 01: Enkapsulasi pada kelas Bank Account

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.__owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"${amount} telah ditambahkan ke akun {self.__owner}. New balance: ${self.__balance}")
        else:
            print("Deposit harus bernilai positif.")

    def withdraw(self, amount):
        if amount > 0:
            if self.__balance >= amount:
                self.__balance -= amount
                print(f"${amount} telah ditarik dari akun {self.__owner}. New balance: ${self.__balance}")
            else:
                print("Saldo tidak mencukupi.")
        else:
            print("Jumlah penarikan harus bernilai positif.")

    def get_balance(self):
        return self.__balance

if __name__ == "__main__":
    alice_account = BankAccount(owner="Alice", balance=1000)

    alice_account.deposit(500)	# Berhasil
    alice_account.deposit(-100)	# Gagal (validasi)

    # Withdraw uang
    alice_account.withdraw(300)	# Berhasil
    alice_account.withdraw(2000)	# Gagal (saldo tidak cukup)

    # Get saldo
    current_balance = alice_account.get_balance()
    print(f"Saldo terakhir di akun {alice_account._BankAccount__owner}: ${current_balance}")

```

### Praktikum 02: Enkapsulasi pada kelas Employee

```python
class Employee:
    def __init__(self, name: str, salary: int):
        self.__name = name
        self.__salary = salary

    def update_salary(self, increase: int):
        if increase > 0:
            self.__salary += increase
            print(f"Gaji {self.__name} telah diperbarui. New salary: ${self.__salary}")
        else:
            print("Kenaikan gaji harus bernilai positif.")
    def set_salary(self, new_salary: int):
        if new_salary > 0:
            self.__salary = new_salary
            print(f"Gaji {self.__name} telah diatur. New salary: ${self.__salary}")
        else:
            print("Gaji harus bernilai positif.")

    def get_salary(self) -> int:
        return self.__salary

    def get_employee_info(self) -> str:
        return f"Employee Name: {self.__name}, Salary: ${self.__salary}"

if __name__ == "__main__":
    employee1 = Employee(name="John Doe", salary=50000)
    print(employee1.get_employee_info())

    employee1.update_salary(5000)
    print(employee1 .get_employee_info())

    employee1.set_salary(60000)
    print(employee1.get_employee_info())
```

### Praktikum 04: Konstruktor dan Destruktor program FileLogger

```python
class FileLogger:
    def __init__(self, filename: str):
        self.__filename = filename
        try:
            self.__file = open(self.__filename, "a")
        except Exception as e:
            print(f"Error: Unable to open file {self.__filename}")
            print(f"File '{filename}' berhasil dibuka untuk logging.")
    def	write_log(self, message):
        """
        Menulis pesan log ke dalam file.

        Parameter:
        - message: Pesan yang akan ditulis ke file.
        """

        self.__file.write(message + "\n")
        self.__file.flush()
        print(f"Pesan log: '{message}' telah ditulis.")


    def __del__(self):
        """
        Destruktor: Menutup file log ketika objek dihapus.
        """
        if hasattr(self, "file") and not self.file.closed: self.file.close()
        print(f"File '{self.__filename}' telah ditutup.")
if  __name__ == "__main__":
    logger = FileLogger("application.log")

    logger.write_log("Aplikasi dimulai.")
    logger.write_log("Melakukan operasi A...")
    logger.write_log("Operasi A selesai.")
    logger.write_log("Aplikasi akan segera selesai.")

    # Menghapus objek logger secara eksplisit
    del logger
```

### Praktikum 05: Properti "self" pada kelas

```python
class Calculator:
    def __init__(self, initial_value: int = 0):
        """
        Konstruktor kelas Calculator.
        -	self: Mengacu pada instance yang dibuat.
        -	initial_value: Nilai awal dari kalkulator.
        """
        self.value = initial_value
        print(f"Kalkulator diinisialisasi dengan nilai: {self.value}")

        self.__value = initial_value

    def add(self, number):
        """
        Menambahkan 'number' ke nilai yang tersimpan di objek.
        -	self.value: Menunjukkan nilai saat ini yang disimpan di objek
        - number: Nilai yang akan ditambahkan. 16:
        """
        self.value += number
        print(f"Setelah penambahan {number}, nilai sekarang adalah: {self.value}")

    def subtract(self, number):
        """
        Mengurangkan 'number' dari nilai yang tersimpan di objek.
        -	self.value: Menunjukkan nilai saat ini yang disimpan di objek
        - number: Nilai yang akan dikurangkan.
        """
        self.value -= number
        print(f"Setelah pengurangan {number}, nilai sekarang adalah: {self.value}")

    def reset(self):
        """
        Mereset nilai kalkulator ke 0.
        """
        self.value = 0
        print(f"Nilai saat ini adalah: {self.value}")

    def show_value(self):
        """
        Menampilkan nilai saat ini dari kalkulator.
        """
        print(f"Nilai saat ini adalah: {self.value}")

def main():
    calc1 = Calculator(initial_value=10)
    calc1.add(5)
    calc1.subtract(3)
    calc1.reset()

    calc2 = Calculator()
    calc2.add(20)
    calc2.show_value()

if __name__ == "__main__":
    main()
```

### Praktikum 06: Menggunakan getter, setter, dan dekorator @property untuk mengelola akses atribut

```python
class Person:
    def __init__(self, name: str, age: int):
        self.__name = name
        self.__age = age

    @property
    def name(self) -> str:
        return self.__name

    @property
    def age(self) -> int:
        return self.__age

    @name.setter
    def name(self, value: str):
        if value:
            self.__name = value
            print(f"Name updated to: {self.__name}")
        else:
            print("Name tidak boleh kosong.")

    @age.setter
    def age(self, value: int):
        if value >= 0:
            self.__age = value
            print(f"Umur updated to: {self.__age}")
        else:
            print("Umur harus bernilai non-negatif.")
def main():
    person = Person("Alice", 30)
    print(f"Name: {person.name}, Umur: {person.age}")

    person.name = "Bob"
    person.age = 25
    print(f"Name: {person.name}, Umur: {person.age}")

    person.age = -5

if __name__ == "__main__":
    main()
```

## Penugasan

```python
class Student:
    def __init__(self, name: str, score: int):
        self.__name = name
        if 0 <= score <= 100:
            self.__score = score
        else:
            print("Score harus bernilai antara 0 dan 100. Score diset ke 0.")
            self.__score = 0
        self.__grade = self.__calculate_grade(self.__score)


    @property
    def name(self) -> str:
        return self.__name

    @property
    def score(self) -> int:
        return self.__score

    def __calculate_grade(self, score: int) -> str:
        if 90 <= score <= 100:
            return "A"
        if 80 <= score <= 89:
            return "B"
        if 70 <= score <= 79:
            return "C"
        if 60 <= score <= 69:
            return "D"
        return "E"

    @name.setter
    def name(self, value: str):
        if value:
            self.__name = value
            print(f"Nama diubah menjadi: {self.__name}")
        else:
            print("Nama tidak boleh kosong.")

    @score.setter
    def score(self, value: int):
        if 0 <= value <= 100:
            self.__score = value
            self.__grade = self.__calculate_grade(value)
            print(f"Nilai diubah...")
        else:
            print("Nilai harus bernilai antara 0 dan 100.")

    def show_info(self):
        print(f"""
Nama Mahasiswa: {self.__name}
Nilai: {self.__score}
Grade: {self.__grade}
        """)

    def __del__(self):
        print(f"Data mahasiswa {self.__name} telah dihapus dari sistem.")

def main():
    student = Student("Siti", 87)
    student.show_info()

    student.score = 90
    student.show_info()

if __name__ == "__main__":
    main()
```
