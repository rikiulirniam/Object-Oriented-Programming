### Praktikum 01: Mendefinisikan Kelas Abstrak Sederhana

```python
from abc import ABC, abstractmethod


class KendaraanAbstrak(ABC):
    def __init__(self, merk):
        self.merk = merk
        print(f"Inisialisasi KendaraanAbstrak dengan merk: {self.merk}")

    def info_merk(self):
        print(f"Merk kendaraan: {self.merk}")

    @abstractmethod
    def start_mesin(self):
        print(f"Mesin mobil {self.merk} dinyalakan.")

    @abstractmethod
    def stop_mesin(self):
        print(f"Mesin mobil {self.merk} dimatikan.")


if __name__== "__main__":
    print("Definisi Kelas Abstrak 'KendaraanAbstrak' selesai.")

    class Mobil(KendaraanAbstrak):
        def start_mesin(self):
            print(f"Mesin mobil {self.merk} dinyalakan.")
        def stop_mesin(self):
            print(f"Mesin mobil {self.merk} dimatikan.")

    print("\nContoh definisi kelas anak 'Mobil' selesai.")
    mobil_contoh = Mobil("Toyota")

    mobil_contoh.start_mesin()
```

### Praktikum 02: Mencoba Instansiasi Kelas Abstrak

```python
from abc import ABC, abstractmethod

class MediaAbstrak(ABC):
    def __init__(self, judul):
        self.judul = judul
        print(f"Inisialisasi MediaAbstrak dengan judul: {self.judul}")

    @abstractmethod
    def play(self):
        pass

    @abstractmethod
    def stop(self):
        pass

if __name__ == "__main__":
    print("Mencoba membuat object dari kelas abstrak 'MediaAbstrak'.")

    try:
    # Baris ini akan menyebabkan TypeError karena MediaAbstrak
    # memiliki metode abstrak (play dan stop) yang belum diimplementasi.
        media = MediaAbstrak("Konten Abstrak")
        # Kode di bawah ini tidak akan pernah dijalankan jika error terjadi
        print("Objek berhasil dibuat (SEHARUSNYA TIDAK TERJADI).")
        media.play()
    except TypeError as e:
    # Menangkap dan menampilkan error TypeError
        print(f"\nGAGAL membuat objek!")
        print(f"Error yang muncul (sesuai harapan): {e}")

        print("\nIni membuktikan bahwa kelas abstrak tidak bisa diinstansiasi")
        print("jika masih memiliki metode abstrak yang belum diimplementasikan.")
```

### Praktikum 03: Membuat Subclass Konkret - Alat Pembayaran

```python
from abc import ABC, abstractmethod
import locale
import random

try:
    locale.setlocale(locale.LC_ALL, 'id_ID.UTF-8')
except locale.Error:
    print("Locale 'id_ID.UTF-8' tidak tersedia. Menggunakan locale default.")

def format_rupiah(jumlah):
    return locale.currency(jumlah, grouping=True, symbol="Rp ")

class AlatPembayaranAbstrak(ABC):
    def __init__(self, nama_metode):
        self.nama_metode = nama_metode
        print(f"Inisialisasi alat pembayaran : {self.nama_metode}")

    def info(self):
        print(f"Metode pembayaran: {self.nama_metode}")

    @abstractmethod
    def proses_pembayaran(self, jumlah):
        """Metode abstrak untuk memproses pembayaran sejumlah 'jumlah'.
        Harus diimplementasikan oleh subclass.
        harus mengembalikan True jika berhasil, False jika gagal. """

class KartuKredit(AlatPembayaranAbstrak):
    def __init__(self, no_kartu, nama_pemilik):
        super().__init__("Kartu Kredit")
        self.no_kartu = no_kartu
        self.nama_pemilik = nama_pemilik
        print(f"Detail Kartu Kredit - No: {self.no_kartu}, Pemilik: {self.nama_pemilik}")

    def proses_pembayaran(self, jumlah):
        print(f" Memproses pembayaran {format_rupiah(jumlah)} menggunakan Kartu Kredit {self.no_kartu} milik {self.nama_pemilik}...")
        berhasil = random.choice([True, False])
        if berhasil:
            print("Pembayaran berhasil!")
            return True
        else:
            print("Pembayaran gagal!")
            return False

class DompetDigital(AlatPembayaranAbstrak):
    def __init__(self, nomor_telepon, nama_provider):
        super().__init__(f"Dompet Digital - {nama_provider}")
        self.nomor_telepon = nomor_telepon
        self.saldo = random.randint(50000, 500000)
        print(f"Detail Dompet Digital - No Telepon: {self.nomor_telepon}, Saldo: {format_rupiah(self.saldo)}")

    def proses_pembayaran(self, jumlah):
        print(f"Memproses pembayaran {format_rupiah(jumlah)} menggunakan Dompet Digital {self.nama_metode} dengan nomor {self.nomor_telepon}...")
        if self.saldo >= jumlah:
            self.saldo -= jumlah
            print(" Pembayaran Dompet Digital Berhasil.")
            print(f"Sisa saldo: {format_rupiah(self.saldo)}")
            return True
        else:
            print("	Pembayaran Dompet Digital Gagal (Saldo tidak mencukupi).")
            return False


if __name__ == "__main__":
    print("\nMembuat Objek Alat Pembayaran...")
    kartu_bca = KartuKredit("1234-5678-9012-3456", "Budi Cahyono")
    gopay = DompetDigital("08123456789", "GoPay")

    print("\nMelakukan Pembayaran:")
    print("\nMencoba bayar dengan Kartu Kredit:")
    kartu_bca.info()
    status_kk = kartu_bca.proses_pembayaran(150000)
    print(f" Status Transaksi KK: {'Sukses' if status_kk else 'Gagal'}")
    print("\nMencoba bayar dengan GoPay (Jumlah Kecil):")
    gopay.info()
    status_gopay1 = gopay.proses_pembayaran(75000)
    print(f" Status Transaksi GoPay 1: {'Sukses' if status_gopay1 else 'Gagal'}")

    print(f" Status Transaksi GoPay 1: {'Sukses' if status_gopay1 else 'Gagal'}")

    print("\nMencoba bayar dengan GoPay (Jumlah Besar):")
    gopay.info()
    status_gopay2 = gopay.proses_pembayaran(1000000) # Kemungkinan gagal karena saldo
    print(f" Status Transaksi GoPay 2: {'Sukses' if status_gopay2 else 'Gagal'}")




```

### Praktikum 04: Menggunakan Kelas Abstrak untuk Polimorfisme - Dokumen

```python
from abc import ABC, abstractmethod
import time

class DokumenAbstrak(ABC):
    def __init__(self, nama_file):
        self.nama_file = nama_file
        print(f"Inisialisasi Dokumen: {self.nama_file}")

    def info_file(self):
        print(f"Nama file: {self.nama_file}")

    @abstractmethod
    def cetak(self):
        """Metode abstrak untuk mencetak dokumen."""
        pass

    @abstractmethod
    def simpan(self):
        """Metode abstrak untuk menyimpan dokumen."""
        pass

class DokumenTeks(DokumenAbstrak):
    def __init__(self, nama_file, isi_teks):
        super().__init__(nama_file)
        self.isi_teks = isi_teks
        print("-> Tipe : Dokumen Teks")

    def cetak(self):
        print(f"Mencetak Dokumen Teks '{self.nama_file}':")
        print("="*15)
        print(self.isi_teks)
        print("="*15)


    def simpan(self):
        print(f"Menyimpan dokumen teks {self.nama_file} ke disk... ")
        time.sleep(0.2)
        print(" -> Berhasil disimpan.")

class Spreadsheet(DokumenAbstrak):
    def __init__(self, nama_file, jumlah_baris, jumlah_kolom):
        super().__init__(nama_file)
        self.jumlah_baris = jumlah_baris
        self.jumlah_kolom = jumlah_kolom
        print("-> Tipe : Spreadsheet ")

    def cetak(self):
        print(f"Mencetak Spreadsheet '{self.nama_file}' ({self.jumlah_baris}baris x {self.jumlah_kolom}kolom):")
        print(" (Menampilkan pratinjau...)")

    def simpan(self):
        print(f"Menyimpan Spreadsheet '{self.nama_file}' ke format .xlsx...")
        # Logika penyimpanan file spreadsheet...
        time.sleep(0.3) # Simulasi
        print(" -> Berhasil disimpan.")

def proses_dokumen(dokumen):
    print("\n======= MEMPROSES SEMUA DOKUMEN =======")
    for doc in dokumen:
        print(f"\n--- Memproses {type(doc).__name__}: {doc.nama_file} ---")
        try:
        # Memanggil metode secara polimorfik
            doc.cetak()
            doc.simpan()
        except Exception as e:
            print(f" Error saat memproses {doc.nama_file}: {e}")
    print("\n======= SELESAI MEMPROSES DOKUMEN =======")


# --- Kode Utama ---
if   __name__	 == "__main__":
    print("Membuat Dokumen...")
    dok1 = DokumenTeks("laporan.txt", "Ini adalah isi laporan singkat.")
    dok2 = Spreadsheet("data_penjualan.xlsx", 200, 15)
    dok3 = DokumenTeks("catatan_rapat.txt", "Poin penting:\n- Bahas budget\n- Tentukan timeline")

    koleksi_dokumen = [dok1, dok2, dok3]

    proses_dokumen(koleksi_dokumen)
    print(f"\nIsi teks dok1: \n{dok1.isi_teks}")
```

### Praktikum 05: Kelas Abstrak dengan Metode Konkret dan Abstrak

```python
from abc import ABC, abstractmethod
import math

class BangunDatarAbstrak(ABC):
    def __init__(self, nama):
        self._nama = nama
        print(f"Inisialisasi Bangun Datar: {self._nama}")

    @property
    def nama_bangun(self):
        return self._nama

    def info_lengkap(self):
        print(f"Bangun Datar: {self.nama_bangun}")

    @abstractmethod
    def hitung_luas(self)-> float:
        pass

    @abstractmethod
    def hitung_keliling(self)-> float:
        pass

class Lingkaran(BangunDatarAbstrak):
    def __init__(self, radius):
        super().__init__("Lingkaran")
        if radius <= 0:
            raise ValueError("Radius harus lebih besar dari nol.")
        self.radius = radius
        print(f"Lingkaran dibuat dgn Radius: {self.radius}")

    def hitung_luas(self) -> float:
        return math.pi * self.radius ** 2

    def hitung_keliling(self) -> float:
        return 2 * math.pi * self.radius

if __name__ == "__main__":
    print("Membuat object lingkaran . . .")
    lingkaran_A = Lingkaran(10)

    print(f"\nNama bangun (via property): {lingkaran_A.nama_bangun}")

    lingkaran_A.info_lengkap()

    luas = lingkaran_A.hitung_luas()
    keliling = lingkaran_A.hitung_keliling()

    print(f"Luas langsung: {luas:.2f}, Keliling langsung: {keliling:.2f}")
```

### Praktikum 06: Properti Abstrak - Elemen Grafis

```python
from abc import ABC, abstractmethod

class ElemenGrafis(ABC):
    def __init__(self, id_elemen, warna):
        self.warna = warna
        self._id = id_elemen

        print(f"[ElemenGrafis __init__] ID: {self.id_elemen}, Warna: {self.warna}")


    @property
    def id_elemen(self):
        return self._id

    @property
    def warna(self):
        return self._warna

    @warna.setter
    def warna(self, nilai):
        self._warna = nilai

    @property
    @abstractmethod
    def posisi(self):
        print("[ElemenGrafis posisi] Mengakses posisi (abstrak) seharusnya di override")
        ...

    @posisi.setter
    @abstractmethod
    def posisi(self, koordinat_baru):
        print("[ElemenGrafis posisi.setter] Mengatur posisi (abstrak) seharusnya di override")
        ...

    @abstractmethod
    def gambar(self):
        print("[ElemenGrafis gambar] Menggambar elemen (abstrak) seharusnya di override")
        ...
        pass



class Kotak(ElemenGrafis):
    def __init__(self, id_elemen, warna, x=0, y=0, lebar=10, tinggi=10):
        super().__init__(id_elemen, warna)

        self._x = x
        self._y = y
        self.lebar = lebar
        self.tinggi = tinggi

    @property
    def posisi(self):
        print(f"Kotak posisi getter dipanggil. Posisi {self.id_elemen} saat ini: ({self._x}, {self._y})")

    @posisi.setter
    def posisi(self, koordinat_baru):
        print(f"Kotak posisis setter dipanggil. Mengubah posisi {self.id_elemen} dari ({self._x}, {self._y}) ke {koordinat_baru}")
        self._x, self._y = koordinat_baru

    def gambar(self):
        print(f"[Kotak gambar] Menggambar Kotak '{self.id_elemen}' warna {self.warna} di	({self._x},{self._y}) dengan ukuran {self.lebar}x{self.tinggi}")

    def info_warna(self):
        print(f"Warna elemen {self.id_elemen}: {self.warna}")

if __name__  == "__main__":
    print("Membuat objek Kotak...")
    kotak1 = Kotak("KotakA", "Merah", x=5, y=10, lebar=20)
    print("-" * 30)


    print("Mengakses ID...")
    id_ktk = kotak1.id_elemen
    print("-" * 30)

	# Mengakses properti abstrak 'posisi' (memanggil getter Kotak)

    print("Mengakses Posisi Awal...")
    pos_awal = kotak1.posisi
    print("-" * 30)

	# Mengubah posisi melalui properti (memanggil setter Kotak) 095:
    print("Mengubah Posisi...")
    kotak1.posisi = (50, 60)
    print("-" * 30)

	# Mengakses posisi lagi
    print("Mengakses Posisi Baru...")
    pos_baru = kotak1.posisi
    print("-" * 30)

	# Mencoba mengubah dengan nilai salah
    print("Mencoba Set Posisi Salah...")
    kotak1.posisi = [100, 200]
    print("-" * 30)

    print("Menggambar Kotak...")
    kotak1.gambar()
    print("-" * 30)

	# Memanggil metode konkret yang diwarisi 115:
    print("Info Warna...")
    kotak1.info_warna()
    print("-" * 30)
```

### Senjata

```python
from abc import ABC, abstractmethod

class Senjata(ABC):
    def __init__(self, nama):
        self.nama = nama
        print(f"Inisialisasi Senjata: {self.nama}")

    @abstractmethod
    def serang(self):
        pass

    @property
    @abstractmethod
    def ammo(self):
        pass

    @ammo.setter
    def ammo(self, nilai):
        pass

    def info_senjata(self):
        print(f"Nama senjata: {self.nama}, Ammo: {self.ammo}")

class Pedang(Senjata):
    def __init__(self, nama, panjang_bilah):
        super().__init__(nama)
        self.panjang_bilah = panjang_bilah
        print(f"Detail Pedang - Panjang Bilah: {self.panjang_bilah} cm")

    def serang(self):
        print(f"{self.nama} menyerang dengan bilah sepanjang {self.panjang_bilah} cm!")

    @property
    def daya_tahan(self):
        return self.panjang_bilah * 10

    @property
    def ammo(self):
        return "Tidak ada ammo, ini adalah senjata melee."

class Panah(Senjata):
    def __init__(self, nama, jumlah_anak_panah):
        super().__init__(nama)
        self.jumlah_anak_panah = jumlah_anak_panah
        print(f"Detail Panah - Jumlah Anak Panah: {self.jumlah_anak_panah}")

    def serang(self):
        if self.jumlah_anak_panah > 0:
            self.jumlah_anak_panah -= 1
            print(f"{self.nama} menembak anak panah! sisa anak panah: {self.jumlah_anak_panah}")
        else:
            print(f"Anak panah {self.nama} habis!")


    @property
    def kecepatan(self):
        return random.randint(1, 10) * 5

    @property
    def ammo(self):
        return self.jumlah_anak_panah

    @ammo.setter
    def ammo(self, nilai):
        self.jumlah_anak_panah = nilai

if __name__ == "__main__":
    print("Membuat objek Pedang...")
    pedang_excalibur = Pedang("Excalibur", 100)
    pedang_excalibur.info_senjata()
    pedang_excalibur.serang()
    print(f"Daya tahan pedang: {pedang_excalibur.daya_tahan}")
    print("-" * 30)

    print("Membuat objek Panah...")
    panah_elven = Panah("Panah Elven", 10)
    panah_elven.info_senjata()
    panah_elven.serang()
    print(f"Kecepatan panah: {panah_elven.kecepatan}")
    panah_elven.serang()
    print(f"Kecepatan panah: {panah_elven.kecepatan}")
    print("-" * 30)

    panah_elven.ammo = 20
    print(f"Menambahkan {panah_elven.ammo} ammo pada Panah...")
    panah_elven.info_senjata()
```
